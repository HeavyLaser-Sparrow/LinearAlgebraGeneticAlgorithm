##Ship Class

```python
scoreIndex = numpy.array([7, 2, 0, 5, 5, 3, 0, 0, 3, 0, 2, 0, 0, 0, 0, 2, 1, 1, 1, 1, 1, 3, 0, 1, 0, 1, 0, 0, 4, 0, 0, 5, 7, 7, 4, 4])
class Ship:
  def __init__(self, name, outfits, vector):
    self.name = name
    self.vector = numpy.array(vector)
    self.outfits = outfits

    # --- 1. EXTRACT RAW STATS ---
    # Based on your baseOutfits, we can infer:
    # Index 6: Mass (Hull strength essentially)
    # Index 7: Drag
    # Index 34: Thrust
    # Index 35: Turn
    # Index 31, 32, 33: Weapon Damages (Shield, Hull, Heat dmg)
    # Index 4: Shield Capacity
    # Index 15: Energy Generation (Reactors usually add to this)
    # Index 22: Energy Capacity (Batteries usually add to this)

    mass = max(1.0, self.vector[6])
    drag = max(0.1, self.vector[7])
    thrust = self.vector[34]

    self.acceleration = thrust / mass
    self.turn = self.vector[35] / drag
    self.maxSpeed = thrust / drag

    # --- 2. CALCULATE COMBAT POWER (MAGNITUDE) ---
    # We reward RAW stats. A bigger number is better.
    # We weight Hull (Mass) and Shields heavily so big ships (Falcon/Dreadnought)
    # automatically beat small ships (Wasp/Interceptor).

    # Sum of all weapon damage types
    total_firepower = self.vector[31] + self.vector[32] + self.vector[33]

    # Survivability = Shield Capacity + Mass (Hull)
    survivability = self.vector[4] + mass

    # --- 3. APPLY SCORING WEIGHTS ---
    # Firepower is king (x10), Survivability is queen (x2), Speed is a bonus (x1000 to scale up small decimals)
    raw_score = (total_firepower * 10) + (survivability * 2) + (self.acceleration * 1000)

    # --- 4. THE SUSTAINABILITY CHECK (CRITICAL FIX) ---
    # If the ship has weapons but no energy capacity/generation, it's a brick.
    # We apply a massive penalty.

    energy_generation = self.vector[15]
    energy_capacity = self.vector[22]

    penalty = 0

    # If we have weapons (firepower > 0) but no energy generation or capacity:
    if total_firepower > 0 and (energy_generation <= 0 and energy_capacity <= 0):
        penalty += 1000000 # Instant fail

    # Heat Management Check (Optional but recommended)
    # If Heat Gen (28) is massive compared to Cooling (29 or similar), penalize.
    if self.vector[28] > 500: # Arbitrary heat limit
        penalty += 50000

    self.score = raw_score - penalty

  # Standard Getters
  def getName(self): return self.name
  def getOutfits(self): return self.outfits
  def getTurn(self): return self.turn
  def getMaxSpeed(self): return self.maxSpeed
  def getAcceleration(self): return self.acceleration
  def getScore(self): return self.score
  def getVector(self): return self.vector
```

```python
weapon_outfits = [
    outfit_name for outfit_name, outfit_vector in baseOutfits.items()
    if outfit_vector[16] > 0 or any(outfit_vector[i] > 0 for i in [31, 32, 33, 34])
]

# Define thruster outfits (positive thrust, element 34)
thruster_outfits = [
    outfit_name for outfit_name, outfit_vector in baseOutfits.items()
    if outfit_vector[34] > 0
]

# Define turn outfits (positive turn, element 35)
turn_outfits = [
    outfit_name for outfit_name, outfit_vector in baseOutfits.items()
    if outfit_vector[35] > 0
]

def is_vector_valid(vector):
  """Helper function to check if a ship's vector is valid."""
  # (Fuel, Outfits, Cargo, Crew capacity must be non-negative)
  if vector[12] < 0 or vector[13] < 0 or vector[14] < 0 or vector[8] < 0:
    return False
  # (Energy capacity must be positive, Heat generation must be less than 50)
  if vector[22] <= 0 or vector[28] >= 50:
    return False
  # New check: Thrust (element 34) and Turn (element 35) must be positive
  if vector[34] <= 0 or vector[35] <= 0:
    return False
  return True

def genShips():
  """Generates a single ship, which may or may not have a weapon or thruster/engine."""
  outfits = []
  newShipName, newShipVector = random.choice(list(baseShips.items()))
  newShipVector = numpy.array(newShipVector)

  # Always add Hyperdrive first
  newShipVector += numpy.array(baseOutfits["Hyperdrive"])
  outfits.append("Hyperdrive")

  # --- Attempt to add at least one weapon ---
  weapon_added = False
  weapon_selection_attempts = 0
  max_weapon_selection_attempts = 50 # Limit attempts

  while not weapon_added and weapon_selection_attempts < max_weapon_selection_attempts:
    random_weapon = random.choice(weapon_outfits)
    testShipVector = newShipVector + numpy.array(baseOutfits[random_weapon])

    # Check if adding this weapon is valid
    if is_vector_valid(testShipVector):
      newShipVector = testShipVector
      outfits.append(random_weapon)
      weapon_added = True
    weapon_selection_attempts += 1

  # --- Attempt to add at least one thruster outfit ---
  thruster_added = False
  thruster_selection_attempts = 0
  max_thruster_selection_attempts = 50 # Limit attempts

  while not thruster_added and thruster_selection_attempts < max_thruster_selection_attempts:
    random_thruster = random.choice(thruster_outfits)
    testShipVector = newShipVector + numpy.array(baseOutfits[random_thruster])

    # Check if adding this thruster is valid (will still check for positive turn, which might not be met yet)
    if is_vector_valid(testShipVector):
      newShipVector = testShipVector
      outfits.append(random_thruster)
      thruster_added = True
    thruster_selection_attempts += 1

  # --- Attempt to add at least one turn outfit ---
  turn_added = False
  turn_selection_attempts = 0
  max_turn_selection_attempts = 50 # Limit attempts

  while not turn_added and turn_selection_attempts < max_turn_selection_attempts:
    random_turn = random.choice(turn_outfits)
    testShipVector = newShipVector + numpy.array(baseOutfits[random_turn])

    # Check if adding this turn outfit is valid (will ensure both thrust and turn are positive now)
    if is_vector_valid(testShipVector):
      newShipVector = testShipVector
      outfits.append(random_turn)
      turn_added = True
    turn_selection_attempts += 1

  # --- Attempt to add other random outfits ---
  testTry = 100
  while testTry > 0:
    addOutfit = random.choice(list(baseOutfits.keys()))

    # Skip if already added
    if addOutfit in outfits:
      testTry -= 1
      continue

    testShipVector = newShipVector + numpy.array(baseOutfits[addOutfit])

    # Only add the outfit if it keeps the ship within valid parameters
    if is_vector_valid(testShipVector):
      outfits.append(addOutfit)
      newShipVector = testShipVector
    testTry -= 1 # Decrement attempt counter

  # Return the newly created ship
  newShip = Ship(newShipName, outfits, newShipVector)
  return newShip
```

## Implement Genetic Algorithm Core Logic


```python
def select_parents(population):
  """Selects the top 50% of the population based on score."""
  population.sort(key=lambda ship: ship.getScore(), reverse=True)
  num_parents = (len(population) + 1) // 2  # Round up for odd sizes
  return population[:num_parents]

def crossover(parent1, parent2, base_ships, base_outfits, weapon_outfits, thruster_outfits, turn_outfits):
  """Combines two parents to create a child ship."""
  # Randomly choose base ship from one of the parents
  chosen_base_ship_name = random.choice([parent1.getName(), parent2.getName()])
  newShipName, newShipVector = chosen_base_ship_name, numpy.array(base_ships[chosen_base_ship_name])

  childOutfits = ["Hyperdrive"]
  newShipVector += numpy.array(base_outfits["Hyperdrive"])

  # Combine outfits from parents, adding only if valid
  all_parent_outfits = list(set(parent1.getOutfits() + parent2.getOutfits()))
  random.shuffle(all_parent_outfits) # Randomize order to prevent bias

  for outfit_name in all_parent_outfits:
    if outfit_name not in childOutfits and outfit_name != "Hyperdrive":
      testShipVector = newShipVector + numpy.array(base_outfits[outfit_name])
      if is_vector_valid(testShipVector):
        newShipVector = testShipVector
        childOutfits.append(outfit_name)

  # Ensure at least one weapon, thruster, and turn outfit
  for outfit_type_list, outfit_type_name in [
      (weapon_outfits, "weapon"),
      (thruster_outfits, "thruster"),
      (turn_outfits, "turn")
  ]:
    has_type = False
    for outfit in childOutfits:
      if outfit in outfit_type_list:
        has_type = True
        break

    if not has_type:
      # Attempt to add a missing type with limited tries
      tries = 0
      max_tries = 20
      while not has_type and tries < max_tries:
        random_outfit = random.choice(outfit_type_list)
        if random_outfit not in childOutfits:
          testShipVector = newShipVector + numpy.array(base_outfits[random_outfit])
          if is_vector_valid(testShipVector):
            newShipVector = testShipVector
            childOutfits.append(random_outfit)
            has_type = True
        tries += 1

  child_ship = Ship(newShipName, childOutfits, newShipVector)
  return child_ship

def mutate(ship, base_outfits, weapon_outfits, thruster_outfits, turn_outfits):
  """Randomly adds or removes an outfit from a ship."""
  mutated_ship = Ship(ship.getName(), list(ship.getOutfits()), numpy.array(ship.getVector()))
  current_outfits = list(mutated_ship.getOutfits())
  current_vector = numpy.array(mutated_ship.getVector())

  action = random.choice(["add", "remove"])

  if action == "add":
    available_outfits = [o for o in base_outfits.keys() if o not in current_outfits]
    if available_outfits:
      outfit_to_add = random.choice(available_outfits)
      testShipVector = current_vector + numpy.array(base_outfits[outfit_to_add])
      if is_vector_valid(testShipVector):
        mutated_ship.outfits.append(outfit_to_add)
        mutated_ship.vector = testShipVector

  elif action == "remove":
    removable_outfits = [o for o in current_outfits if o != "Hyperdrive"]

    if removable_outfits:
      random.shuffle(removable_outfits) # Randomize selection for removal
      for outfit_to_remove in removable_outfits:
        testShipVector = current_vector - numpy.array(base_outfits[outfit_to_remove])

        # Check if removing would leave the ship without required outfits
        temp_outfits = [o for o in current_outfits if o != outfit_to_remove]
        has_weapon_after_remove = any(o in weapon_outfits for o in temp_outfits)
        has_thruster_after_remove = any(o in thruster_outfits for o in temp_outfits)
        has_turn_after_remove = any(o in turn_outfits for o in temp_outfits)

        if (is_vector_valid(testShipVector) and
            has_weapon_after_remove and
            has_thruster_after_remove and
            has_turn_after_remove):

          mutated_ship.outfits.remove(outfit_to_remove)
          mutated_ship.vector = testShipVector
          break # Successfully removed one outfit, exit loop

  # Re-calculate score, maxSpeed, acceleration, and turn for the mutated ship
  # This is important because the vector changed and these are derived attributes
  mutated_ship.turn = mutated_ship.vector[35] / mutated_ship.vector[6] # thrust/drag
  mutated_ship.maxSpeed = mutated_ship.vector[35] / mutated_ship.vector[7] # thrust/mass
  mutated_ship.acceleration = mutated_ship.vector[34] / mutated_ship.vector[6]
  vectorNorm = numpy.linalg.norm(numpy.array(mutated_ship.vector)) #norm of vector ||v||
  if vectorNorm != 0:
    normalVector = numpy.array(mutated_ship.vector) / vectorNorm #normalized vector v/||v||
    mutated_ship.score = numpy.dot(normalVector, scoreIndex)
  else:
    mutated_ship.score = 0.0

  return mutated_ship
```

## Run Genetic Algorithm Simulation


```python
current_population = list(myShips)
initial_population_size = len(myShips)
generations = 1000
mutation_rate = 0.10 # 10% mutation rate

average_scores = []

print("Starting Genetic Algorithm simulation...")

for generation in range(generations):
  # 5. Select parents
  selected_parents = select_parents(current_population)

  # 6. Create new population with selected parents
  new_population = list(selected_parents)

  # 7. Fill new population with children generated by crossover and mutation
  while len(new_population) < initial_population_size:
    # Randomly select two parents for crossover
    parent1 = random.choice(selected_parents)
    parent2 = random.choice(selected_parents)

    # 8. Perform crossover
    child_ship = crossover(parent1, parent2, baseShips, baseOutfits, weapon_outfits, thruster_outfits, turn_outfits)

    # 9. Apply mutation
    if random.random() < mutation_rate:
      child_ship = mutate(child_ship, baseOutfits, weapon_outfits, thruster_outfits, turn_outfits)

    # 10. Add child to new population
    new_population.append(child_ship)

  # 11. Update current population
  current_population = new_population

  # 12. Calculate average score and append to list
  gen_scores = [ship.getScore() for ship in current_population]
  avg_score = sum(gen_scores) / len(gen_scores)
  average_scores.append(avg_score)

  # 13. Print progress update
  if (generation + 1) % 50 == 0 or generation == 0 or generation == generations - 1:
    print(f"Generation {generation + 1}/{generations}, Average Score: {avg_score:.2f}")

print("Genetic Algorithm simulation complete.")
```
